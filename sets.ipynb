```python
num = {1,2,3,3,2,2,4,4,5,6,6,9}
print(num)
#set allow only unique values
```

```python
# we can convert alist of dupilicate values to set to get unique values
num = [1,1,1,12,3,4,3,2,2,2,4,4,5,6,6,9]
num = set(num)
print(num)
```

```python
# even if we want to uodate a dupilicate value it will not add same extra value
nums = {1,2,3,4,5}
nums.add(3)
print(nums)
nums.add(6)
print(nums)
```

```python
print(nums)
nums.update([1,2,3,4,5,6,7,8,9,9])
print(nums)
```

```python
nums.remove(9)
print(nums)
```

```python
nums.discard(10) # discard will not throw error if the value is not present in the set
print(nums)
nums.remove(10) #remove will throw error if the value is not present in the set 
print(nums)
```

```python
nums.pop() # pop removes the first element from the set but since set is unordered we cannot predict which element will be removed
print(nums)
```

```python
s1 = {1,2,3,4,3,True,7}
s2 = {3,4,7,8,3,False,9}

print(s1.union(s2))
```

```python
print(s1.intersection(s2))
```

```python
print(s1.difference(s2))
```

```python
print(s1.symmetric_difference(s2))
```
