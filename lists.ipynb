```python
#we can add data to the list using append() method, by using this method we can add only one element at a time
nums=[1,2,3,4,5]
print(nums)
nums.append(6)
print(nums)
```

```python
# we can add more elements then the append() method using extend() method
nums.extend([7,8,9,10])
print(nums)
```

```python
#we can count the no of times a element is present in the list using the coount () method
print(nums.count(5))
```

```python
# we can get the index of the element using the index () method
print(nums.index(1))
print(nums.index(10))
```

```python
# find, we dont have find () method in the list like we have in the strings, but we have a diff approach to use find method 
print(10 in nums)
print(11 in nums)
```

```python
# in list we can perform the CRUD operations
# we can delete a element from the list using the pop() method, by default it will dlete thelast element in the list
# but we can de;ete a specific element by using the index of the element 
nums.pop()
print(nums)
nums.pop(0)
print(nums)
```

```python
# we can insert a element in the list using the insert() methond
# by the index we can insert a element where ever we want
nums.insert(0,1)
print(nums)
```

```python
num = [5,2,7,5,8,4,7]
num.sort()
print(num)
num.sort(reverse=True)
print(num)
# we can use the sort method to sort the elements in the list in asecening or decending order
```

```python
# we can copy elements in one list to the other list using the copy() method
num1= num.copy()
print(num1)
```
