```python
#comprehension is a way to create a new list by applying an expression to each item in an existing iterable (like a list, tuple, or string). It allows for concise and readable code.
```

```python
nums = [ x for x in range(1,11)]
print(nums)
even = [ x for x in nums if x%2==0]
print(even)
```

```python
#ternary operator : it is a way to write an if else statement in a single line.
a=10
b=20
c=30
d=40

big = a if a>b and a >c and a >d else b if b>c and b>d else c if c>d else d
print(big)
```

```python
val = [val%2==0 for val in range(1,11)]
print(val)
```

```python
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
# print(matrix)
vals = [ e for r in matrix for e in r]
print(vals)
```

```python
squares = {x : x**2 for x in range(1,11)}
print(squares)
```

```python
squares = {x : x%2==0 for x in range(1,11)}
print(squares)
```

```python
squares = {x : x%2==0 for x in range(1,11) if x== 5}
print(squares)
```

```python
details = {
    "hemanth" : True,
    "mahi" : False,
    "honey" : True,
    "om" : False
}
passed = {k:v for k,v in details.items() if v==True}
print(passed)
vow = {k:v for k,v in details.items() if k[0] in "aeiouAEIOU"}
print(vow)
```
