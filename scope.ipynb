```python
#scope of a variable is the regiomm of the program where the variable is accessible.
#there are 4 types of scopes :
#local scope, ecnclosed scope, global scope, builtin scope
```

```python
num =10 
def var():
    num = 20
    print(num)
var()
print(num)
```

```python
def sample():
    name = "hemanth" #local scope variable
    print(name)
sample()
print(name) # name is not defined because it is in local scope and cannot be accessed outside of the function.
```

```python
n = 10 #global scope variable
def v1():
    print(n)
def v2():
    print(n)
v1()
v2()
# we can access the glocal scope variable from any function in the program.
```

```python
#access and modifying the global variable inside a function
n = 0
def inc():
    global n # we need to use the global keyword to modify the global vraiable inside the function.
    n+=1
    print(n)
def dec():
    global n
    n-=1
    print(n)
inc()
inc()
dec()
```

```python
def user():
    global name
    name= input("enter your name : ")
    print(name)
def is_valid():
    print(isinstance(name,str))
user()
is_valid()
#we can define the global variable insife the function and access it in other functions as well.
```

```python
#enclosed scope is the scope of a variable that is defined inside a function and can be accessed by the inner function.
n = 10 
def outter():
    def inner():
        print(n)
    inner()
outter()
```

```python
n = 10 
def outter():
    n=100
    def inner():
        print(n)
    inner()
outter()
```

```python
n = 10 
def outter():
    n=100
    def inner():
        n = 1000
        print(n)
    inner()
outter()
```

```python
# built in scope is the scope of a variable that is defined inside a function and can be accessed by the inner function.
def outter():
    def inner():
        print(len("hello"))
    inner()
outter()
```

```python
count = 0
def is_even(arr):
    global count
    for num in arr:
        if num%2==0:
            count+=1
    return count
print(is_even([1,2,3,4,5,6]))
    
```
