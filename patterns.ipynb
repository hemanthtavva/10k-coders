```python
# star pyramid pattern
rows =5
for i in range (1, rows+1):
    res = " "
    for j in range (1, rows+1):
        if j <= rows-i:
            res += "  "
        else:
            res += "* "
    print(res)
```

```python
#pyramid pattern
rows = 5
for i in range (1, rows+1):
    res="" 
    for sp in range (1,( rows-i)+1):
        res+=" " 
    for j in range (1, i+1):
        res+= "* "
    print(res)
```

```python
#inverted pyramid pattern
rows= 5
for i in range(rows, 0, -1):
    res=""
    for sp in range(1, rows - i + 1):
        res+= " "
    for j in range (1, i+1):
        res+= "* "
    print(res)  
```

```python
# I pattern
rows =5
for i in range (1, rows+1):
    res="" 
    for j in range (1, rows+1):
        if i ==1 or i == rows or j ==(rows):
            res += "* "
        else :
            res += " "
    print(res)
```

```python
# H pattern
rows = 5
for i in range (1, rows+1):
    res=""
    for j in range (1, rows+1):
        if j==1 or j == rows or i == (rows//2)+1:
            res += "* "
        else:
            res += "  " 
    print(res)
```

```python
# N Pattern
rows = 5
for i in range (1, rows+1):
    res=""
    for j in range (1, rows+1):
        if j ==1 or j == rows or i == j:
            res += "*"
        else :
            res += " "
        
    print(res)
```

```python
# z pattern
rows = 5
for i in range (1, rows+1):
    res=""
    for j in range (1, rows+1):
        if i == 1 or i == rows or i+j == rows+1:
            res += "*"
        else :
            res += " "
        
    print(res)
```

```python
# T pattern
rows = 5
for i in range (1, rows+1):
    res=""
    for j in range (1, rows+1):
        if i == 1 or j == (rows//2)+1: 
            res += "*"
        else :
            res += " "
        
    print(res)
```

```python
#s pattern 
rows =5
mid = rows//2 +1
for i in range (1, rows+1):
    res="" 
    for j in range (1, rows+1):
        if i ==1 or j == mid or i == rows or (j == 1 and i <mid) or ( i > mid and j == rows):
            res += "* "
        else :
            res += "  "
    print(res)
```

```python
#pattern for the multiplies 
rows = 5
for i in range(1, rows+1):
    res=""
    for j in range (1, rows+1):
        res += str(i*j) + " "
    print(res)
    
```

```python
#rhombus pattern
rows = 5
for i in range (1, rows+1):
    res="" 
    for sp in range (1,( rows-i)+1):
        res+=" " 
    for j in range (1, i+1):
        res+= "* "
    print(res)

for i in range( rows, 0 ,-1):
    res="" 
    for sp in range (1, (rows - i) +1):
        res+= " "
    for j in range (1, i+1):
        res+= "* "
    print(res)
```

```python
# hour glass pattern 
rows= 5
for i in range(rows, 0, -1):
    res=""
    for sp in range(1, rows - i + 1):
        res+= " "
    for j in range (1, i+1):
        res+= "* "
    print(res)   

for i in range (2,rows+1):
    res=""
    for sp in range(1, (rows - i)+1):
        res+= " "
    for j in range (1, i+1):
        res+="* "
    print(res)
```

```python
# X O X O X
# O X O X O
# X O X O X
# O X O X O
# X O X O X
rows= 5 
for i in range(1, rows+1):
    res=""
    for j in range(1, rows+1):
        if (i+j) % 2 == 0:
            res += "X "
        else:
            res += "O "
    print(res)
```
