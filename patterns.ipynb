```python
# star pyramid
rows =5
for i in range (1, rows+1):
    res = " "
for sp in range (1, rows - i+1):
    res += " "
    for j in range (1, j+1):
        res += "* "
print(res)
```

```python
rows = 5
for i in range (1, rows+1):
    res="" 
    for j in range (1, rows+1):
        if i == j or i == rows or i ==(rows//3) +1 or j=1:
            res += "* "
        print(res)
```

```python
rows = 5
for i in range (1, rows+1):
    res=""
    for j in range (1, rows+1):
        if j==1 or j == rows or j == (rows//2)+1:
            res += "* "
        else:
            res += "  "
        print(res)
```
