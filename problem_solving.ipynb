```python
#anagaram check
s1 = "pot"
s2 = "top"
is_anagram = True
if len(s1) == len(s2):
    for char in s1 :
        if s1.count(char) != s2.count(char):
            is_anagram= False
            break
else :
    is_anagram = False

print(is_anagram)
```

```python
#anagram check using sort
s1 = "now"
s2 = "won"
s1= list(s1)
s1.sort()
s1 = ''.join(s1)
s2 = list(s2)
s2.sort()
s2=''.join(s2)
print(s1==s2)
```

```python
#anagaram check using sorted 
s1 = "now"
s2 = "won"
print(sorted(s2))
print(sorted(s1)==sorted(s2))
```

```python
#string rotation
name = "Hemanth"
rotations =427
if rotations%len(name)==0:
    print(name)
else:
    rotations = rotations%len(name)
    first = name[:rotations]
    second = name[rotations:]
    print(second + first)
```

```python
#accesing the json data
details = [
    {"name":"Hemanth", "marks":[50,70,100]},
    {"name":"mahi", "marks":[90,70,100]},
    {"name":"Hema", "marks":[50,80,100]}

]
stu_name = "" 
max_marks = 0
for stu in details :
    total_marks = sum(stu["marks"])
    if total_marks > max_marks:
        max_marks = total_marks
        stu_name = stu["name"]
print(stu_name)
```

```python
#accesing the json data and finding the best subject
details = [
    {"name":"Hemanth", "marks":{"tel":50,"hin":70,"eng":100}},
    {"name":"mahi", "marks":{"tel":90,"hin":70,"eng":97}},
    {"name":"Hema", "marks":{"tel":50,"hin":80,"eng":90}}

]
for stu in details:
   sub=""
   mark = 0
   for key in stu["marks"]:
      if stu["marks"][key]> mark:
        mark = stu["marks"][key]
        sub = key
        stu["best_subject"] = sub
print(details)


```

```python
#finding the toppers in each subject
details = [
    {"name":"Hemanth", "marks":{"tel":50,"hin":70,"eng":100}},
    {"name":"mahi", "marks":{"tel":90,"hin":70,"eng":97}},
    {"name":"Hema", "marks":{"tel":50,"hin":80,"eng":90}}

]
report={}
tel = 0
hin = 0
eng = 0
for stu in details:
    for key in stu["marks"]:
        if key == "tel":
            if stu["marks"][key]>tel:
                tel = stu["marks"][key]
                report[key] = stu["name"] 

        if key == "hin":
            if stu["marks"][key]>hin:
                hin = stu["marks"][key]
                report[key] = stu["name"] 

        if key == "eng":
            if stu["marks"][key]>eng:
                eng = stu["marks"][key]
                report[key] = stu["name"] 
print(report)
```
