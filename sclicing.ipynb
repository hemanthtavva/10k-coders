```python
s = "hemanth Tavva"
sub = "Tavva"

exist = False
for i in range(len(s)):
    if s[i] == sub[0]:
        temp = s[i: i+len(sub)]
        if temp == sub:
            exist = True
            break
print(exist)
#explain this code line by line:
# 1. `s = "hemanth Tavva"`: This line initializes a string variable `s` with the value "hemanth Tavva".
# 2. `sub = "Tavva"`: This line initializes another string variable `sub` with the value "Tavva".
# 3. `exist = False`: This line initializes a boolean variable `exist` and sets it to `False`. This variable will be used to indicate whether the substring `sub` is found in the string `s`.
# 4. `for i in range(len(s)):`: This line starts a for loop that iterates over the indices of the string `s`. The loop will run from `0` to `len(s) - 1`. This allows us to check each character in the string `s` one by one.
# 5. `if s[i] == sub[0]:`: Inside the loop, this line checks if the current character `s[i]` is equal to the first character of the substring `sub[0]`. If this condition is true, it means we have found a potential starting point for the substring `sub` in the string `s`.           
# 6. `temp = s[i: i+len(sub)]`: If the previous condition is true, this line creates a new string `temp` that contains a slice of the string `s` starting from index `i` and ending at index `i + len(sub)`. This slice is intended to capture the portion of the string `s` that is the same length as the substring `sub`.
# 7. `if temp == sub:`: This line checks if the sliced string `temp` is equal to the substring `sub`. If this condition is true, it means that we have found the substring `sub` in the string `s`.
# 8. `exist = True`: If the previous condition is true, this line sets the variable `exist` to `True`, indicating that the substring `sub` has been found in the string `s`.
# 9. `break`: This line exits the for loop immediately after finding the substring `sub`, since we no longer need to continue searching through the string `s`.
# 10. `print(exist)`: After the loop has completed, this line prints the value of the variable `exist`, which will be `True` if the substring `sub` was found in the string `s`, and
```

```python
word = "abcdecfghitjmvrxaz"
longest=""
for i in range(len(word)):
    temp = ""
    for j in range(i, len(word)):
        if word[j] not in temp:
            temp += word[j]
        else:
            break
        if len(temp) > len(longest):
            longest = temp
print(longest)
```
