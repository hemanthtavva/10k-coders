```python
word = " hello world" # olleh dlrow
word = word.split(" ")
for words in range (len(word)):
    #print(word[words])
    rev =""
    for i in range (len(word[words])-1,-1,-1):
        rev += word[words][i]
    word[words]= rev
print(" ".join(word))
    
```

```python
word = " hello world" # olleh dlrow using slicing
word = word.split(" ")
for words in range (len(word)):
    word[words] = word[words][::-1]
print(" ".join(word))
```

```python
#ASCII -> American Standard Code for Information Interchange

# ord - order, returns the ASCII value of a character
# chr - charcter, returns the character of a given ASCII value

print(ord("a")) #97-122
print(ord("A")) #65-90

print(ord("0")) #48-57
print(ord("7"))

print(ord("+"))
print(ord("-"))
print(ord("*"))
```

```python
print(chr(97))
print(chr(122))
print(chr(65))
print(chr(57))
print(chr(48))
```

```python
#in the name if any vowel is present then we have to replace that vowel with the next character in alphabets
name = "hemanth" #hfmbnth
vowels = "aeiou"
res =""
for i in name:
    if i in vowels:
        code = ord(i)
        code += 1
        res += chr(code)
    else:
        res += i
print(res)
```
