```python
# dictonary is a data type, it stores data in key value pair, it is mutable, it is unordered, it is indexed, it is denoted by {}
```

```python
details = { "name" : "Hemanth tavva",
            "age" : 21,
            "city" : "hyd",
            "skills" : ["python", "data science", "sql"],
            "address" : {"street" : "123 main st", "city" : "hyd", "state" : "telangana"}
            }
print(details)
print(type(details))
print(details["name"])
```

```python
#CRUD operations on dictionary
#create
details["email"] = "hemanthtavva7@gmail.com"
print(details)
#read
print(details["email"])
#update
details["age"] = 22
print(details)
#delete
del details["email"]
print(details)
```

```python
for a in details:
    print(details[a])
```

```python
for key, value in details.items():
    print(key,":" ,value)
```

```python
#keys() it returns all the keys in the dictonary
print(details.keys())
```

```python
#values() it returns all the values in the dictonary
print(details.values())
```

```python
#items() it returns all the key value pairs in the dictonary
print(details.items())
```

```python
# use case : remove a key value pair from the dictonary
updated_details= {}
for k,v in details.items():
    if k != "employes":
        updated_details[k] = v
print(updated_details)
```

```python
#get() it returns the value of the specified key
print(details.get("gender")) # it returns none if the key is not present in the dictonary
print(details.get("gender", "male")) # it returns the defalut value if the key is not present in the dictonary
```

```python
#pop() it removes the specified key and returns the corresponding value
print(details)
print(details.pop("skills"))
```

```python
print(details["address"]["street"])
```
